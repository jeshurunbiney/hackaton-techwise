```python
from PIL import Image, ImageDraw, ImageFont
import textwrap
import io

# Access the Image Path
image_path = list(uploaded.keys())[0]

def create_meme(image_path, top_text, bottom_text, font_path='/usr/share/fonts/truetype/liberation/LiberationSans-Bold.ttf', font_size=80, output_path='meme_output.jpg'):
    # Load the image using PIL
    img = Image.open(io.BytesIO(uploaded[image_path]))
    draw = ImageDraw.Draw(img)
    image_width, image_height = img.size

    # Load a TrueType font
    font = ImageFont.truetype(font_path, font_size)

    # Define text position and size
    top_text = textwrap.fill(top_text, width=15)
    bottom_text = textwrap.fill(bottom_text, width=15)

    # Function to draw text with stroke
    def draw_text_with_stroke(draw, text, position, font, text_color='white', stroke_color='black', stroke_width=4):
        x, y = position
        # Draw stroke
        for dx in range(-stroke_width, stroke_width + 1):
            for dy in range(-stroke_width, stroke_width + 1):
                if dx != 0 or dy != 0:
                    draw.text((x + dx, y + dy), text, font=font, fill=stroke_color)
        # Draw text
        draw.text((x, y), text, font=font, fill=text_color)

    # Calculate text size using textbbox
    top_text_bbox = draw.textbbox((0, 0), top_text, font=font)
    bottom_text_bbox = draw.textbbox((0, 0), bottom_text, font=font)

    top_text_width = top_text_bbox[2] - top_text_bbox[0]
    top_text_height = top_text_bbox[3] - top_text_bbox[1]
    bottom_text_width = bottom_text_bbox[2] - bottom_text_bbox[0]
    bottom_text_height = bottom_text_bbox[3] - bottom_text_bbox[1]

    # Calculate top text position (25% vertically centered, horizontally right-aligned)
    top_text_x = (image_width - top_text_width) / 1.2  # Adjust to the right
    top_text_y = image_height * 0.25 - top_text_height / 2

    # Calculate bottom text position (75% vertically centered, horizontally right-aligned)
    bottom_text_x = (image_width - bottom_text_width) / 1.2  # Adjust to the right
    bottom_text_y = image_height * 0.75 - bottom_text_height / 2

    # Draw text on the image with stroke
    draw_text_with_stroke(draw, top_text, (top_text_x, top_text_y), font)
    draw_text_with_stroke(draw, bottom_text, (bottom_text_x, bottom_text_y), font)

    # Save the image
    img.save(output_path)

# Example usage
create_meme(image_path, 'MAKING GAMES FOR HACKATON', "MAKING MEMES")
```

```python
from PIL import Image, ImageDraw, ImageFont
import textwrap
import io

# Access the Image Path
image_path = list(uploaded.keys())[0]

def create_meme(image_path, top_text, bottom_text, font_size= 10000, output_path='meme_output.jpg'):
    # Load the image using PIL
    img = Image.open(io.BytesIO(uploaded[image_path]))
    draw = ImageDraw.Draw(img)
    image_width, image_height = img.size

    # Load a built-in font
    font = ImageFont.load_default()

    # Define text position and size
    top_text = textwrap.fill(top_text, width=150)
    bottom_text = textwrap.fill(bottom_text, width=150)

    # Function to draw text with stroke
    def draw_text_with_stroke(draw, text, position, font, text_color='white', stroke_color='black', stroke_width=2):
        x, y = position
        # Draw stroke
        for dx in range(-stroke_width, stroke_width + 1):
            for dy in range(-stroke_width, stroke_width + 1):
                if dx != 0 or dy != 0:
                    draw.text((x + dx, y + dy), text, font=font, fill=stroke_color)
        # Draw text
        draw.text((x, y), text, font=font, fill=text_color)

    # Calculate text size using textbbox
    top_text_bbox = draw.textbbox((0, 0), top_text, font=font)
    bottom_text_bbox = draw.textbbox((0, 0), bottom_text, font=font)

    top_text_width = top_text_bbox[2] - top_text_bbox[0]
    top_text_height = top_text_bbox[3] - top_text_bbox[1]
    bottom_text_width = bottom_text_bbox[2] - bottom_text_bbox[0]
    bottom_text_height = bottom_text_bbox[3] - bottom_text_bbox[1]

    # Calculate top text position (25% vertically centered, horizontally centered)
    top_text_x = (image_width - top_text_width) / 2
    top_text_y = image_height * 0.25 - top_text_height / 2

    # Calculate bottom text position (75% vertically centered, horizontally centered)
    bottom_text_x = (image_width - bottom_text_width) / 2
    bottom_text_y = image_height * 0.75 - bottom_text_height / 2

    # Draw text on the image with stroke
    draw_text_with_stroke(draw, top_text, (top_text_x, top_text_y), font)
    draw_text_with_stroke(draw, bottom_text, (bottom_text_x, bottom_text_y), font)

    # Save the image
    img.save(output_path)

# Example usage
create_meme(image_path, 'MAKING GAMES FOR HACKATON', "GENERATING MEMES")
```
